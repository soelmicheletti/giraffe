## 1. Loading data

```python
import giraffe
import matplotlib.pyplot as plt
from netZooPy.otter.otter import otter
from netZooPy.panda import Panda
import notebooks
import numpy as np
import pandas as pd
from data.prostate import preprocessing
from sklearn.metrics import roc_curve, auc
import time
```

```python
cache = True
```

```python
if not cache:
    preprocessing.generate_data()
else:
    expression = pd.read_csv("data/prostate/raw/expression.csv", index_col = 0)
    motif = pd.read_csv("data/prostate/motif.csv", index_col = 0)
    ppi = pd.read_csv("data/prostate/ppi_matrix.csv", index_col = 0)
    chip = pd.read_csv("data/prostate/chip.csv", index_col = 0)
```

```python
print(expression.shape)
print(motif.shape)
print(ppi.shape)
print(chip.shape)
```

## 2. Benchmarking

### 2.1 GIRAFFE

```python
if cache:
    R_giraffe = pd.read_csv("data/prostate/R_giraffe.csv", index_col = 0)
else:
    start = time.time()
    R_giraffe = giraffe.Giraffe(expression.to_numpy(), motif.to_numpy(), ppi.to_numpy() + np.eye(ppi.shape[0]), save_computation = True).get_regulation()
    end = time.time()
    print(end - start)
    pd.DataFrame(R_giraffe, index = motif.index, columns = motif.columns).to_csv("data/prostate/R_giraffe.csv")
```

### 2.2 OTTER [2]

```python
if cache:
    R_otter = pd.read_csv("data/prostate/R_otter.csv", index_col = 0)
else:
    start = time.time()
    R_otter = otter(motif.T.to_numpy(), ppi.to_numpy() + np.eye(ppi.shape[0]), np.corrcoef(expression.to_numpy())).T
    end = time.time()
    print(end - start)
    pd.DataFrame(R_otter, index = motif.index, columns = motif.columns).to_csv("data/colon/R_otter.csv")
```

### 2.3 PANDA [1]

```python
R_panda_raw = pd.read_csv("data/prostate/panda.csv", index_col = 0).T
R_panda = pd.DataFrame(0, index = motif.index, columns = motif.columns)
for g in R_panda_raw.index:
    for tf in R_panda_raw.columns:
        R_panda.at[g, tf] = R_panda_raw.at[g, tf]
```

### 2.4 Motif-based prior

```python
notebooks.plot_auroc(chip.to_numpy(), [motif.to_numpy().flatten()], ["Motif"])
```

### 2.5 GENIE3 [3]

```python
model = notebooks.GENIE3(expression.to_numpy().T, gene_names = list(expression.index) ,regulators = list(expression.index), ntrees=5)
```

```python
genie_full = pd.DataFrame(model, index = motif.index, columns = motif.index)
R_genie = pd.DataFrame(0, index=motif.index, columns= motif.columns)
translate = pd.read_csv('data/prostate/raw/gen_v26_mapping.csv')
for tf in motif.columns:
    transl = list(translate[translate['gene_name'] == tf]['gene_id'])
    for ens in transl:
        if ens[0:15] in motif.index:
            R_genie.loc[:, tf] += genie_full.loc[:, ens[0:15]]
```

### 2.6 TIGRESS [4]

```python
translation = pd.read_csv("data/prostate/raw/gen_v26_mapping.csv")
tf_ids = {}
for tf in motif.columns:
    trans = list(translation[translation['gene_name'] == tf]['gene_id'])
    if len(trans) == 0:
        continue
    else:
        trans = trans[0][0:15]
        tf_ids[trans] = tf
R_tigress = pd.DataFrame(0, index=motif.index, columns=motif.columns)
files = ["R_tigress_2", "R_tigress_5002", "R_tigress_10002", "R_tigress_15002", "R_tigress_20002", "R_tigress_25002"]
for file in files:
    a = pd.read_csv("data/prostate/" + file + ".csv", index_col = 0)
    for tf in a.index:
        for g in a.columns:
            if isinstance(a.at[tf, g], pd.Series):
                continue
            if a.at[tf, g] == 0:
                continue
            if tf not in tf_ids.keys():
                continue
            R_tigress.at[g, tf_ids[tf]] = a.at[tf, g]
    print("File done!")
```

### 2.7 BITFAM [5]

```python
translation = pd.read_csv("data/prostate/raw/gen_v26_mapping.csv")
tf_ids = {}
for tf in motif.columns:
    trans = list(translation[translation['gene_name'] == tf]['gene_id'])
    if len(trans) == 0:
        continue
    else:
        trans = trans[0][0:15]
        tf_ids[trans] = tf
R_bitfam = pd.DataFrame(0, index=motif.index, columns=motif.columns)
files = ['1', '3001', '6001', '9001', '12001', '15001', '18001', '21001', '24001', '27001']
for file in files:
    a = pd.read_csv("data/prostate/R_bitfam_" + file + ".csv", index_col = 0)
    for tf in a.columns:
        for g in a.index:
            if tf not in tf_ids.keys():
                continue
            R_bitfam.at[g, tf_ids[tf]] = a.at[g, tf]
    print("File done!")
```
