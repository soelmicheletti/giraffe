```python
import mygene
import numpy as np
import pandas as pd
```

```python
expr_lcss = pd.read_csv("raw/rna_expression_norm_lccs.txt", sep = "\t")
```

```python
mg = mygene.MyGeneInfo()
M = {}
for gene in expr_lcss[expr_lcss.columns[0]]:
    print(gene)
    result = mg.query(gene, scopes="symbol", fields=["ensembl"], species="human", verbose=False)
    hgnc_name = gene
    for hit in result["hits"]:
        if "ensembl" in hit and "gene" in hit["ensembl"]:
            M[gene] = hit["ensembl"]["gene"]

translation = []
for i in range(expr_lcss.shape[0]):
    sym = expr_lcss.iloc[i, 0]
    if sym in M.keys():
        translation.append(M[sym])
    else:
        translation.append("None")
expr_lcss["ENS"] = translation
```

```python
expr = pd.DataFrame((expr_lcss.iloc[:, 2:(expr_lcss.shape[1]-1)]).to_numpy(), index = expr_lcss["ENS"])
expr = expr.sort_index()
expr = expr.drop(["None"], axis = 0)
```
